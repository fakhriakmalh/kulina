```python
# import seaborn
import seaborn as sns
# settings for seaborn plotting style
sns.set(color_codes=True)
# settings for seaborn plot sizes
sns.set(rc={'figure.figsize':(5,5)})
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('D:/DATA SCIENCE/cari sendiri/Preliminary-Test-master/Statistics/dataset.csv')
user = data['user_id']


#Plot Data

fig, ax = plt.subplots()
ax.set(xlabel='Normal distribution', ylabel='Frequency')
sns.distplot(user, bins=25, color="g", ax=ax)
plt.show()
```

the graph that i use above was the continuous probability distributions, which can tell the frequency of user_id in a certain range. The higher value of frequency means the higher probability of the user to use the Apps. 

As you can see, that the distribution of the user data is the left-skewed distribution, so the distribution which can suit the data is the left-skewed distribution. As we can see that the probability of user_id in range 75000 until 120000 is higher than other id. 

Based on the graph, we can tell that the user_id in range 75000 until 120000 is the more loyal customer than others. We should check the gender and age of the user so we can make a better customer segmentation to boost the business.
